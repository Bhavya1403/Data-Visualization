```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the JSON file into a DataFrame
df = pd.read_json('../Databases/documentDB.json')
```

```python
# Visualization 1: Price distribution
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
# Visualization 2: Ratings distribution
plt.figure(figsize=(10, 6))
plt.hist(df['rating'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution')
plt.show()
```

```python
# Pie chart for display distribution
plt.figure(figsize=(10, 8))
display_counts = df['display'].value_counts()
plt.pie(display_counts, labels=display_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Display Distribution')
plt.axis('equal')
plt.show()
```

```python
# Convert battery life to numeric values
# Extracting numerical values from the 'batteryLife' column and converting them to hours
df['batteryLife_hours'] = df['batteryLife'].str.extract(r'(\d+)', expand=False).astype(float)

# Bar chart for battery life across different smartphone models
plt.figure(figsize=(12, 8))
plt.bar(df['model'], df['batteryLife_hours'], color='blue')
plt.xlabel('Smartphone Model')
plt.ylabel('Battery Life (hours)')
plt.title('Battery Life Across Smartphone Models')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
